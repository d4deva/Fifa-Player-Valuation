```python
from bs4 import BeautifulSoup
from urllib.request import urlopen as ureq
import time
import os
import requests
import numpy as np
import pandas as pd
```

```python
pages=range(1,101)
```

```python
Webpage_2=[]
for page in pages:
    Webpage_1 = requests.get("https://www.fifaindex.com/players/{}/".format(page)).text
    page_soup = BeautifulSoup(Webpage_1,'lxml')
        
    table = page_soup.find_all("table",{"class":"table table-striped table-players"})

    for tr in table:
        table_row = tr.tbody.find_all("tr")
        player = tr.find_all("td",{"data-title":"Name"})
        
    for link in player:
        Webpage_2.append("https://www.fifaindex.com"+str(link.find('a')['href']))
```
